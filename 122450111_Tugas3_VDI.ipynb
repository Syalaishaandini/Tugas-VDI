```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
data['Extracurricular_Activities'] = labelencoder.fit_transform(data['Extracurricular_Activities'])
data['Gender'] = labelencoder.fit_transform(data['Gender'])
data.head(10)
```

```python
urutan_kategori = ['Low', 'Medium', 'High']
pemetaan_ordinal = {kategori: urutan for urutan, kategori in enumerate(urutan_kategori)}

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(urutan_kategori)

data['Teacher_Quality'] = data['Teacher_Quality'].apply(lambda x: pemetaan_ordinal.get(x))
data['Access_to_Resources'] = data['Access_to_Resources'].apply(lambda x: pemetaan_ordinal.get(x))

data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Gender', y='Exam_Score', data=data)
plt.title('Perbandingan Nilai Rata-rata Berdasarkan Gender')
plt.show()
```

```python
sns.boxplot(x='Extracurricular_Activities', y='Exam_Score', data=data)
plt.title('Pengaruh Aktivitas Ekstrakurikuler terhadap Nilai')
plt.show()
```

```python
sns.scatterplot(x='Access_to_Resources', y='Exam_Score', data=data)
plt.title('Hubungan Antara Akses ke Sumber Daya dan Nilai')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih kolom numerik saja
numeric_data = data.select_dtypes(include=['number'])

# Menghitung matriks korelasi
correlation_matrix = numeric_data.corr()

# Membuat heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")

plt.title('Korelasi Antar Kolom Data Numerik')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.show()
```
